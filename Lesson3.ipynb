```python
# Задание 3.4.1
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return (x ** 2 - 1 - y), (np.exp(x) + x - x * y - 1)

x = np.linspace(-2, 5, 200)
x1 = fsolve(f, (-2, -1))
x2= fsolve(f, (2, 3))
x3= fsolve(f, (4, 5))
print(x1)
print(x2)
print(x3)
plt.plot(x, x ** 2 - 1)
plt.plot(x, (np.exp(x)-1)/x + 1)
plt.ylim(-1, 15)
plt.grid(True)
plt.show()
```

```python
# Задание 3.4.2
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return (x ** 2 - 1 - y), (np.exp(x) + x - x * y - 1)

def f2(a, b):
    x = fsolve(f, (a, b))
    if f3(x[0], x[1]) > 0:
        print(x)
    
def f3(a, b):
    return np.exp(a) + a - a * b - 1

x = np.linspace(-2, 5, 200)
f2(-2, 1)
f2(2, 3)
f2(4, 5)

plt.plot(x, x ** 2 - 1)
plt.plot(x, (np.exp(x)-1)/x + 1)
plt.ylim(-1, 15)
plt.grid(True)

plt.show()
```
