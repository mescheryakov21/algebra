```python
#Задание 2.3.1
# Окружность
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1, 1, 200)
y = np.sqrt(1 - x**2)
y1 = -(np.sqrt(1 - x**2))
plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel("ось X")
plt.ylabel("ось Y")
plt.axis('equal')
plt.show()
```

```python
#Задание 2.3.2
# Эллипс
x = np.linspace(-2, 2, 200)
a = 2
b = 1
y = np.sqrt(pow(1/b, 2) - pow(x/(a * b), 2))
y1 = - (np.sqrt(pow(1/b, 2) - pow(x/(a * b), 2)))
plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel("ось X")
plt.ylabel("ось Y")
plt.axis('equal')
plt.show()
```

```python
#Задание 2.3.3
# Гипербола
x = np.linspace(-4, 4, 400)
a = 2
b = 1
y = np.sqrt((x / (a * b)) ** 2 - (1 / b) ** 2)
y1 = -(np.sqrt((x / (a * b)) ** 2 - (1 / b) ** 2))

plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel("ось X")
plt.ylabel("ось Y")
plt.show()
```
